````python
# This is the class definition/ structure for the flight booking system


# I am adding some functions outside so as to not confuse with existing intent present in the Intent-Classification
# Here each intent is actually a sub-classification of each intent (such as booking-ticket). So using seperate functions will be appropriate

class FlightBookingSystem:
    def __init__(self, user_input):
        self.user_input = user_input

    def classify_intent(self):
        if "booking" in self.user_input and self.user_input["booking"]:
            return self.handle_booking(self.user_input['booking'][0])

        elif "cancelling" in self.user_input and self.user_input["cancelling"]:
            return self.handle_cancellation(self.user_input['cancelling'][0])

        elif "flight_info" in self.user_input and self.user_input["flight_info"]:
            return self.handle_flight_info(self.user_input['flight_info'][0])

        elif "change_flight" in self.user_input and self.user_input["change_flight"]:
            return self.handle_change_flight(self.user_input['change_flight'][0])

        elif "change_preferences" in self.user_input and self.user_input["change_preferences"]:
            return self.handle_change_preferences(self.user_input['change_preferences'][0])

        else:
            return {'intent': 'unknown'}

    def handle_booking(self, booking_info):
        while not booking_info['Name']:
            booking_info['Name'] = input("What is your name for the booking? ")
        if booking_info['class'] == "not-applicable":
            booking_info['class'] = input("What class do you prefer (economy/business)? ")
        if booking_info['seat'] == "not-applicable":
            booking_info['seat'] = input("What is your seat preference? ")
        if booking_info['source'] == "not-applicable":
            booking_info['source'] = input("What is your departure city? ")
        if booking_info['destination'] == "not-applicable":
            booking_info['destination'] = input("What is your destination city? ")
        if booking_info['meal'] == "not-applicable":
            booking_info['meal'] = input("What is your meal preference? ")

        return {'intent': 'book_flight', 'booking': booking_info}

    def handle_cancellation(self, cancellation_info):
        while not cancellation_info['Name']:
            cancellation_info['Name'] = input("What is your name for the cancellation? ")
        if cancellation_info['flight_number'] == "not-applicable":
            cancellation_info['flight_number'] = input("What is your flight number? ")
        if cancellation_info['reason'] == "not-applicable":
            cancellation_info['reason'] = input("What is the reason for cancellation? ")
        if cancellation_info['pnr_number'] == "not-applicable":
            cancellation_info['pnr_number'] = input("What is your PNR number? ")

        return {'intent': 'cancel_flight', 'cancelling': cancellation_info}

    def handle_flight_info(self, flight_info_request):
        while not flight_info_request['Name']:
            flight_info_request['Name'] = input("What is your name for the flight information request? ")
        if flight_info_request['flight_number'] == "not-applicable":
            flight_info_request['flight_number'] = input("What is your flight number? ")
        if flight_info_request['date_of_travel'] == "not-applicable":
            flight_info_request['date_of_travel'] = input("What is the date of your travel? ")

        return {'intent': 'request_flight_info', 'flight_info': flight_info_request}

    def handle_change_flight(self, change_flight_info):
        while not change_flight_info['Name']:
            change_flight_info['Name'] = input("What is your name for the flight change request? ")
        if change_flight_info['flight_number'] == "not-applicable":
            change_flight_info['flight_number'] = input("What is your current flight number? ")
        if change_flight_info['new_date'] == "not-applicable":
            change_flight_info['new_date'] = input("What is your new date of travel? ")
        if change_flight_info['new_time'] == "not-applicable":
            change_flight_info['new_time'] = input("What is your new time of travel? ")

        return {'intent': 'change_flight', 'change_flight': change_flight_info}

    def handle_change_preferences(self, preferences_info):
        while not preferences_info['Name']:
            preferences_info['Name'] = input("What is your name for the preference change request? ")
        if preferences_info['flight_number'] == "not-applicable":
            preferences_info['flight_number'] = input("What is your flight number? ")
        if preferences_info['class'] == "not-applicable":
            preferences_info['class'] = input("What is your new class preference (economy/business)? ")
        if preferences_info['seat'] == "not-applicable":
            preferences_info['seat'] = input("What is your new seat preference? ")
        if preferences_info['meal'] == "not-applicable":
            preferences_info['meal'] = input("What is your meal preference? ")

        return {'intent': 'change_preferences', 'change_preferences': preferences_info}
````
